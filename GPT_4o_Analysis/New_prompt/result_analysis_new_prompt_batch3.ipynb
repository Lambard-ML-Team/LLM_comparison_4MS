```python
question_template= """
<question>
{question}
</question>

<correct_answer>
{correct_answer}
</correct_answer>

<predicted_answer>
{predicted_answer}
</predicted_answer>

<question_type>
{question_type}
</question_type>
"""
```

```python
def prompt_template(question, correct_answer, predicted_answer, question_type):
    template = question_template.format(
        question=question,
        correct_answer=correct_answer,
        predicted_answer=predicted_answer,
        question_type=question_type
    )
    return template
```

```python
from openai import OpenAI

client = OpenAI(
  api_key=os.environ['OPENAI_API_KEY'],  # this is also the default, it can be omitted
)

# Function to use the OpenAI API
def comparison(questions):
    """
    Ask the model to compare the answer provided by the model to the correct answer.

    Parameters:
    questions (list): A list of questions asked to the model.

    Returns:
    list: A list of answers provided by the model.
    """

    prompt = """
    You are an AI judge tasked with evaluating the correctness of a predicted answer to a given question. Your goal is to determine if the predicted answer is equivalent to the correct answer and if the reasoning behind the predicted answer is sound.

    You will be provided with the following information:
    <question>
    {question}
    </question>

    <correct_answer>
    {correct_answer}
    </correct_answer>

    <predicted_answer>
    {predicted_answer}
    </predicted_answer>

    <question_type>
    {question_type}
    </question_type>

    The question type will be one of the following: MCQ (Multiple Choice Question), MATCH (Match the Following), MCQN (Multiple Choice Numerical Question), or NUM (Numerical Question).

    To evaluate the predicted answer, follow these steps:

    1. Carefully read the question, correct answer, and predicted answer.

    2. Based on the question type, use the following criteria:
    - For MCQ: Check if the predicted answer matches the correct option(s). If multiple options are correct, ensure all are included in the predicted answer.
    - For MATCH: Verify that the predicted answer contains the correct set of matched entities.
    - For MCQN: Confirm that the predicted answer matches the correct numerical option and that the reasoning is sound.
    - For NUM: Compare the predicted numerical answer to the correct answer, allowing for minor rounding differences as specified in the question.

    3. Evaluate the reasoning behind the predicted answer, if provided. Ensure it is logically sound and relevant to the question.

    4. In your internal thought process, consider the following:
    - Is the predicted answer equivalent to the correct answer?
    - Is the reasoning (if provided) correct and relevant?
    - Are there any discrepancies or errors in the predicted answer?
    - For numerical questions, is the answer within an acceptable range of the correct answer?

    5. Based on your evaluation, prepare a brief explanation of your judgment. This explanation should be concise but clear, highlighting the key factors that influenced your decision.

    6. Provide your final judgment and explanation in the following format:
    <judgment>
    Explanation: [Your brief explanation here]
    Result: [CORRECT or INCORRECT]
    </judgment>

    Remember, a predicted answer should only be judged as correct if it is equal or very close to the correct answer AND the reasoning (if provided) is correct. Be thorough in your evaluation and clear in your explanation.
    """
    answers = []
    Token = []
    for question in questions:
        response = client.chat.completions.create(
            model="gpt-4o",
            temperature=0,
            messages=[
                {"role": "assistant", "content":prompt},
                {"role": "user", "content": question}
            ]
        )
        # Extract the response from the model
        model_response = response.choices[0].message.content
        answers.append(model_response)
        Token.append(response.usage)
    return answers,Token
```

```python
test = 0
a = full_answer.sort_values(['Question Info'],ascending=True,ignore_index=True)

for i in correct_data_list:
    data_files_list = gb.glob(i+'*')
    # collecting all the answer in different dataframe
    model=model_list[test]
    Ans_model =[]
    file_name = []
    topic_2 = []
    for j in data_files_list:
        folder_path = j
        for filename in gb.glob(os.path.join(folder_path, '*.txt')):
            with open(filename, 'r') as f:
                text = f.read()
                Ans_model.append(text)
                file_name.append(f.name.strip(j)[:-3])
                topic_2.append(j.strip(i))
    Ans_model = pd.DataFrame(Ans_model,columns=[f'{model}_answer'])
    file_name = pd.DataFrame(file_name,columns=['Question Info_2'])
    topics = pd.DataFrame(topic_2,columns=["TOPIC_2"])
    # Dataframe of the answer provided by the model
    Model = pd.concat((topics,file_name,Ans_model),axis=1)
    # Aligning the question number for the original file (a) and the answer from model (b)
    b = Model.sort_values(['Question Info_2'],ascending=True,ignore_index=True)
    check = pd.concat((a,b),axis=1) #dataframe containing question, correect answer and model answer
    questions = []
    for l in range(len(full_answer)):
        question = str({check['QUESTION'].iloc[l][1:-1]})
        correct_answer = str({check["Correct Answer"].iloc[l]})
        predicted_answer = str({check[f'{model}_answer'].iloc[l]})
        question_type = str({check['Question Type'].iloc[l]})
        goal = prompt_template(question, correct_answer, predicted_answer, question_type)
        questions.append(goal)
    answers, token_usage = comparison(questions)
    result = pd.concat((check,pd.DataFrame(answers,columns=[f"Analysis Result {model}"]),pd.DataFrame(token_usage,columns=["Output_token","Input_token","Total_token"])),axis=1)
    result.to_csv(f"result_{model}.csv")
    test +=1
```
